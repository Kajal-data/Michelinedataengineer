# Create Data from differnt sources

## Create data from file formats

### Create data from csv

```python
dfr = spark.read
df = dfr.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/dataset-20260601T053440Z-3-001/dataset/sample_dataset/users_001.csv",
header = True,
inferSchema = True,
)
df.display()
```

### Create DataFrame from Custom Delimiter

```python
dfr = spark.read
df = dfr.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/dataset-20260601T053440Z-3-001/dataset/sample_dataset/users_001.dat",
header = True,
inferSchema = True,
)
df.display()
```

```python
dfr = spark.read
df = dfr.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/dataset-20260601T053440Z-3-001/dataset/sample_dataset/users_001.dat",
header = True,
inferSchema = True,
sep='|',
)
df.display()
```

### Create DataFrame from Parquet

```python
dfr = spark.read
df = dfr.parquet("/Volumes/quickstart_catalog/quickstart_schema/sendbox/dataset-20260601T053440Z-3-001/dataset/sample_dataset/users.parquet",
header = True,
inferSchema = True,
)
df.display()
```

### Create DataFrame from JSON(Java script object notation)

```python

df = spark.read.json(path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/dataset-20260601T053440Z-3-001/dataset/sample_dataset/users_001.json")
df.display()
```
