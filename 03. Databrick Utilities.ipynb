```python
dbutils.widgets.combobox(
    name="top_n_records",
    defaultValue="5",
    choices=["5", "10", "15", "20"],
    label="Top n Records",
)
```

```python
dbutils.widgets.dropdown(
    name="environment",
    defaultValue="Development",
    choices=["Development", "QA Environment", "Pre Production Environment"],
    label="ENvironment Type",
)
```

```python
dbutils.widgets.text(name="name", defaultValue="Anonymous", label="Enter name")
dbutils.widgets.multiselect(name="technology", defaultValue="2G", choices=["2G","3G","4G","5G"], label="Technology")
```
