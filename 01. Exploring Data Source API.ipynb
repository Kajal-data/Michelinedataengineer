# Custom Schema

```python
from pyspark.sql.types import *
UserSchema = StructType(
    [
        StructField("id", IntegerType()),
        StructField("age", IntegerType()),
        StructField("gender", StringType()),
        StructField("designation", StringType()),
        StructField("salary", IntegerType()),
    ]
)
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/dataset-20260601T053440Z-3-001/dataset/sample_dataset/users_001.csv",
    header=True,
)
df.display()
```

# Handling Bad Records

1. PERMISSIVE
2. DROPMALFORMED
3. FAILFAST

## 1. PERMISSIVE (Default)

```python
df = spark.read.json(path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/dataset-20260601T053440Z-3-001/dataset/access_logs.json").display()
```

```python
df = (
    spark.read.option("ColumnNameOfCorruptRecord", "Bad_Record")
    .json(
        path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/dataset-20260601T053440Z-3-001/dataset/access_logs.json"
    )
    .display()
)
```

# DataFrameWriter API

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/dataset-20260601T053440Z-3-001/dataset/sample_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
df.write.save(
    path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/output/output.json1",
    format="JSON",
)
spark.read.json(
    path="/Volumes/quickstart_catalog/quickstart_schema/sendbox/output/output.json1",
).display()
```
